# use thread 

```python
from threading import Thread
from time import time
```

def get_infobox_thread(name,d):
    d[name]=get_infobox(name)

```python
n=0
def get_infobox_thread(name,d):
    d[name]=get_infobox(name)
    global n
    n+=1
    if not n%100:
        print(n)
```

```python
t0 = time()
dth={}
threadlist=[]
for i in missing['ISSUER_NAME']:
    t = Thread(target=get_infobox_thread,args=(i,dth))
    t.start()
    threadlist.append(t)
for i in threadlist:
    i.join()
t1 = time(); d1 = t1 - t0
```

```python
d1
```

```python
dnew=popNone(dth)
```

```python
table=dictToTable(dnew)
```

# concurrent.futures.ThreadPoolExecutor 

```python
from concurrent.futures import ThreadPoolExecutor
```

```python
t0 = time()
dth={}
executor = ThreadPoolExecutor(100)
for i in missing['ISSUER_NAME']:
    executor.submit(get_infobox_thread, i,dth)
t1 = time(); d1 = t1 - t0
```

```python
d1
```

```python
table = dictToTable(dth)
```
