```python
from web_spider import get_infobox
import pandas as pd
from time import time
from concurrent.futures import ThreadPoolExecutor
```

```python
def dictToTable(d):
    dnew = {key:val for key,val in d.items() if val!=None }
    table = pd.DataFrame.from_dict(dnew,orient='index').reset_index()
    table = table.rename(columns={'index':'Company'})
    return table
```

```python
n=0
def get_infobox_thread(name,d):
    d[name]=get_infobox(name)
    global n
    n+=1
    if not n%100:
        print(n)
```

```python
data = pd.read_pickle("quandl_ib_missing_dict.pk")
company = list(data.values())
company[:10]
```

```python
len(company)
```

```python
t0 = time()
dth={}
executor = ThreadPoolExecutor(1000)
for i in company:
    executor.submit(get_infobox_thread, i,dth)
t1 = time(); d1 = t1 - t0
```
